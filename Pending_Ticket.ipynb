```python
dfa.index = np.arange(1,len(dfa)+1)
dfa
```

```python
#print(dfa.columns)
ta=dfa['Submission time']
inta=ta.index
```

```python
ageing=[];
now = datetime.datetime.now() 
format = '%Y-%m-%d'
for i in range(len(inta)):
  s1=ta[inta[i]]
  s2,s3=s1.split(" ");
  datetime_str = datetime.datetime.strptime(s2, format)
  diff=(now.date()-datetime_str.date()).days
  if(diff==0):
     diff=1;
  ageing.append(diff);
```

```python
u=0;
for index, row in dfa.iterrows():
    dfa.at[index,'Pending Days']= ageing[u]
    u=u+1;
catg=dfa['Category'].tolist();
```

```python
uniq_age=[];
uniq_categ=[];
for i in ageing:
  if i not in uniq_age:
        uniq_age.append(i);
for i in catg:
  if i not in uniq_categ:
        uniq_categ.append(i);
uniq_age.sort();
```

```python
tota=[];
for i in range(len(uniq_age)):
    v=0;
    for j in range(len(ageing)):
       if(ageing[j]==uniq_age[i]):
          v=v+1;
    tota.append(v);
```

```python
vala=[];
for i in uniq_categ:
    valb=[];
    for j in range(len(uniq_age)):
      v=0;
      for k in range(len(ageing)):
        if(ageing[k]==uniq_age[j] and catg[k]==i):
          v=v+1;
      valb.append(v);
    vala.append(valb);
vala.append(tota);
```

```python
for i in range(len(vala)):
   v=0;
   for j in range(len(vala[i])):
     if(vala[i][j]!=" "):
        v=v+vala[i][j];
   vala[i].append(v);
uniq_categ.append('Total');
uniq_age.append('Total');
```

```python
new_df = pd.DataFrame(vala, columns = uniq_age,index=uniq_categ)
new_df
```

```python
new_df.drop(["Total"], inplace = True)
new_df.sort_values(by=['Total'],ascending=False, inplace=True)
new_df
```

```python
s10=name.split("-")
s10=s10[-1].split(".")
s10=s10[0];
if(len(s10)==9 and s10[0]==" "):
   s1=s10[1:3];
   s2=s10[3:5];
   s3=s10[5:9];
   format="%d-%m-%Y"
   s=s1+'-'+s2+'-'+s3;
   dat = datetime.datetime.strptime(s, format)
   dat=dat.date()-datetime.timedelta(days=1);
   dat=str(dat);
   s1,s2,s3=dat.split('-');
   s2=int(s2);
   s2=calendar.month_abbr[s2]
   file_name=s3+s2+s1;
   file_name='Pending ticket as on '+file_name+'.xlsx';
else:
    file_name='Pending ticket.xlsx';
file_name
```

```python
if(os.path.exists(file_name)):
   os.remove(file_name)
writer = pd.ExcelWriter(file_name,engine='xlsxwriter')
#all_tick.to_excel(writer, 'all_tickets')
dfa.to_excel(writer, 'extracted_tickets')
new_df.to_excel(writer, 'Summary',startrow=2)
rows = len(new_df.axes[0]);
cols = len(new_df.axes[1]);
workbook  = writer.book
merge_format = workbook.add_format({
    'bold': 1,
    'border': 1,
    'align': 'center',
    'valign': 'vcenter',
    'fg_color': '#99CCFF'})
no_bold = workbook.add_format({
    'bold': 1,
    'border': False,
    'align': 'center',
    'valign': 'vcenter',
    'fg_color': '#99CCFF'})
border_format = workbook.add_format({
    'border': 1})
bold_format = workbook.add_format({
    'bold': 1,
    'border': 1})
no_bold_format = workbook.add_format({
    'bold': False,
    'border': 1})
worksheet = writer.sheets['Summary']
worksheet.set_column('A:A', 35)
s2='B2:'+c+'2';
worksheet.merge_range(s2,'Ageing in Days',merge_format)
s2=d+'2:'+d+'3';
worksheet.merge_range(s2,'Total',merge_format)
worksheet.merge_range('A2:A3','Applications',merge_format)
#worksheet.merge_range(s10,merge_format)
s2=3;
s1=ord('A');
k=s2+rows;
for j in range(cols):
  s1=s1+1;
b1='A2';
b2='A'+str(k);
b5='A4';
b6='A'+str(k-1);
b3=chr(s1)+str(2);
b4=chr(s1)+str(k);
worksheet.conditional_format( b1+':'+b4 , { 'type' : 'no_blanks' , 'format' : bold_format} )
ch=chr(ord('A')+len(new_df.index));
s15=ch+str(4)+":"+ch+str(3+len(new_df.columns));
s16=ch+str(4)+":"+ch+str(2+len(new_df.columns));
worksheet.conditional_format( s15 , { 'type' : 'no_blanks' , 'format' : border_format} )
worksheet.conditional_format( s16 , { 'type' : 'no_blanks' , 'format' : no_bold} )
bold = workbook.add_format({'bold': True,'align': 'center','valign': 'vcenter'})
worksheet = writer.sheets['extracted_tickets']
worksheet.write('A1', 'S.No', bold)
writer.save()
new_df
```

```python
import openpyxl
from openpyxl.utils.cell import get_column_letter
wb = openpyxl.load_workbook(file_name)
wb.sheetnames
sheet = wb['Summary']
number_rows = len(new_df.index)+5;
number_columns =len(new_df.columns)+5;
for i in range(number_columns):
    for k in range(number_rows):
        cell = str(sheet[get_column_letter(i+1)+str(k+1)].value)
        if(cell=='0'):
            sheet[get_column_letter(i+1)+str(k+1)]="-";
wb.save(file_name)
```

```python
from openpyxl.styles import Font
from openpyxl.styles import PatternFill
wb = openpyxl.load_workbook(file_name)
wb.sheetnames
sheet = wb['Summary']
for i in range(inds,inde+1,1):
  for j in col_y:
    c=sheet.cell(row=i,column=j);
    if(c.value!="-"):
      c.fill = PatternFill(fill_type='solid', start_color='FFA500', end_color='FFA500')
for i in range(inds,inde+1,1):
  for j in col_r:
    c=sheet.cell(row=i,column=j);
    if(c.value!="-"):
      c.fill = PatternFill(fill_type='solid', start_color='FF1111', end_color='FF1111')
cols=2;
cole=len(new_df.columns);
for j in range(cols,cole+1,1):
  c=sheet.cell(row=3,column=j);
  c.fill = PatternFill(fill_type='solid', start_color='99CCFF', end_color='99CCFF')
wb.save(file_name)
print("File Saved Succesfully!!!");
```
